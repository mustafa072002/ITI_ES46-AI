```python
class Product:
    def __init__(self , name = "", price = 0 , quantity = 0):
        self.name = name
        self.price = price
        self.quantity = quantity
        
        print(f"{self.name} Added with {self.quantity} -- Price = {self.price}")
        pass
    
    def restock(self,amount):
        self.quantity += amount
        print(f"{self.name} Restocked with {self.quantity} -- Price = {self.price}")
        
    def sell(self , amount):
        if amount <= self.quantity:
            self.quantity -= amount
            print(f"Selled {self.name} {amount}")
            
        else:
            print("Stock Isn't enough")
            
            
    
```

```python
Apple = Product("Apple" , 10 , 50)

Apple.sell(30)
Apple.sell(20)
```

```python
class Calculator:
    def __init__(self):
        pass
    
    def Add(self,*args):
        sum = 0
        for x in args:
            sum += x
            
        return sum
    
    def Sub(self,*args):
        sum = 0
        for x in args:
            sum -= x
            
        return sum
    
    def Mul(self,*args):
        sum = 1
        for x in args:
            sum *= x
            
        return sum
    
    def Div(self,*args):
        sum = 0
        for x in args:
            if x == 0 :
                print("Divison By Zero Math ERROR")
                return
            else:
                sum += x
            
        return sum


Calc = Calculator()

Calc.Add(10 , 15 ,20)
Calc.Sub(10 , 15 ,20)

    
    
```

```python
import random
class BankSys:
    def __init__(self , Name  , firstDeposit):
        self.Name = Name
        self.AccountNumber = random.randint(0,100)
        self.Balance = firstDeposit
        
    def Withdraw(self , amount):
        if(amount <= self.Balance):
            self.Balance -= amount
            print(f"Withdrawed {amount} , Remaining Balance = {self.Balance}")
            
        else:
            print("Balance Not Enough")
            
    def Deposit(self , amount):
        self.Balance += amount
        print(f"Deposited {amount} , Remaining Balance = {self.Balance}")
        
    def ShowBalance(self):
        print(f"Remaining Balance = {self.Balance}")
        
        
myAccount = BankSys("Mustafa Mahgoub" , 1000)
myAccount.ShowBalance()

myAccount.Deposit(5000)
myAccount.ShowBalance()

myAccount.Withdraw(5000)
myAccount.Withdraw(8000)
        
```

```python
class Book:
    def __init__(self , title , author):
        self.title = title
        self.author = author
        
        
class Library:
    def __init__(self):
        self.library = []
        
    def addBook(self , *book):
        for bk in book:
            self.library.append(bk)
            
    def listBook(self):
        print(f"Available Books {self.library}")
        
        
book1 = Book("HarryPotter1" , "RGG")
book2 = Book("HarryPotter2" , "RGG")

Lib = Library()
Lib.addBook(book1,book2)
Lib.listBook()
```
